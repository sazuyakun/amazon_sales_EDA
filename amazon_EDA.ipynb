## Importing the libraries
---

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('amazon.csv')
```

```python
df.head()
```

```python
df.info()
```

## Data Cleaning
---

```python
df.columns
```

```python
df['discounted_price'] = df['discounted_price'].str.replace('₹', '')
df['actual_price'] = df['actual_price'].str.replace('₹', '')
df['discount_percentage'] = df['discount_percentage'].str.replace('%','')
df['discounted_price'] = df['discounted_price'].str.replace(',','')
df['actual_price'] = df['actual_price'].str.replace(',','')
df['rating_count'] = df['rating_count'].str.replace(',','')
df[(df['rating'] == '|')] = df.replace('|','')
df[(df['rating'] == '')] = df.replace('','4.0')
# df.replace('','4.0') --> check later
```

```python
# Converting the strings to float types
df['discounted_price']= df['discounted_price'].astype(float)
df['actual_price']= df['actual_price'].astype(float)
df['discount_percentage']= df['discount_percentage'].astype(int)
df['rating_count'] = df['rating_count'].astype(float)
df['rating']= df['rating'].astype(float)
```

### Datatype Information

```python
df.info()
```

## Removing the unnecessary columns
---

```python
df = df.drop(columns=['product_link', 'img_link'])
```

```python
df.head()
```

# Analysis
---

```python
df.isnull().sum()
```

```python
# Replacing null values with mean value
rating_mean = np.floor(df['rating_count'].mean())
df['rating_count'].fillna(value = rating_mean, inplace=True)
rating_mean
```

```python
df['rating_count'].isnull().sum()
```

## BARPLOTS
---

```python
sorted_df = df.sort_values(by='rating_count', ascending=False)
top_10 = sorted_df.iloc[:10, :]

plt.figure(figsize=(20,8), dpi=200)
sns.barplot(top_10, x='category', y='rating_count')
```

### Barplot - Categories by rating_count:

Rating Count: Total purchases assumed by rating count.
Category: Top 10 purchase choices according to product category.
This graph indicates that the top 10 purchase categories belong to the electronics category.

---

```python
sorted_df_rating = df.sort_values(by='rating', ascending=False)
top_10_rating = sorted_df_rating.iloc[:10, :]

plt.figure(figsize=(20,25), dpi=200)
sns.barplot(top_10_rating,x='rating',y='product_name')
```

### Barplot - Rating by products

This graph displays the top 20 ratings on the website. The majority of the top 20 ratings belong to electronics and computer accessories, falling between 4 and 5.

---

```python
sorted_df_dis = df.sort_values(by='discount_percentage', ascending=False)
top_10_dis = sorted_df_dis.head(20)
sns.set(rc={'figure.figsize':(20,15)})
sns.barplot(top_10_dis,x='discount_percentage',y='category')
```

### Barplot - Discount percentage by product category.

This graph displays the top 20 discount percentage.The discount percentage is more than 80%. In the top 20 categories, the maximum discount belongs to electronics and computer accessories.

---

```python
sorted_df_dis_p = df.sort_values(by='discounted_price', ascending=False)
top_10_dis_p = sorted_df_dis_p.head(20)
sns.set(rc={'figure.figsize':(20,15)})
sns.barplot(top_10_dis_p,x='discounted_price',y='user_name')
```

### Barplot - Discounted_price by user_name.

This graph displays the top 20 discounted prices. Among these 20 discounted prices, the maximum purchasing capacity belongs to the price range between Rs. 30,000 and Rs. 50,000.

## Conculsion

Conclusion: The barplots offer a comprehensive glimpse into the dynamics of the online marketplace.

The first graph, "Categories by rating_count," depicts the total purchases based on rating count, unveiling the top 10 purchase choices categorized by product type. Notably, the electronics category dominates these choices, suggesting a strong preference for electronic products among consumers.

Next, the "Rating by products" graph showcases the top 20 ratings on the website, with electronics and computer accessories garnering the majority of these high ratings, particularly falling between 4 and 5. This underscores the popularity and satisfaction associated with these product categories.

In the third graph, "Discount percentage by product category," the top 20 discount percentages are displayed, with discounts exceeding 80%. Remarkably, electronics and computer accessories emerge as the categories offering the highest discounts, indicating lucrative savings opportunities for consumers.

Finally, the "Discounted_price by user_name" graph highlights the top 20 discounted prices, revealing that the maximum purchasing capacity lies within the price range between Rs. 30,000 and Rs. 50,000. This suggests a significant interest among consumers in mid to high-priced products, particularly in this price range.

Together, these barplots paint a vivid picture of consumer behavior, preferences, and market trends, guiding both consumers and sellers in navigating the dynamic landscape of online commerce.
